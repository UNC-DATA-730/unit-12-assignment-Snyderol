**Create and visualize your model below...**

````r
set.seed(123)
auto_split = initial_split(Auto, prop = 0.8)
auto_train = training(auto_split)
auto_test = testing(auto_split)
````

````r
auto_recipe = recipe(mpg ~ displacement + cylinders, data = auto_train) |>
  step_ns(displacement, deg_free = 4) |>
  step_mutate(cyl_cat = factor(cylinders)) |>
  step_dummy(cyl_cat) |>
  step_rm(cylinders)
````

````r
auto_model = linear_reg() |>
  set_engine("lm")
````

````r
auto_workflow = workflow() |>
  add_recipe(auto_recipe) |>
  add_model(auto_model)
````

````r
mod_fit = fit(auto_workflow, data = auto_train)
````

````r
auto_preds = predict(mod_fit, auto_test) |>
  bind_cols(auto_test)
````

````r
ggplot(auto_preds, aes(x = displacement, y = mpg)) +
  geom_point(size = 1, alpha = 0.35) +
  geom_line(aes(y = .pred, color = factor(cylinders)), linewidth = 1) +
  labs(title = "MPG as a function of Displacement and Cylinders",
       x = "Displacement",
       y = "MPG")
````
